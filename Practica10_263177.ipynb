Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
```

Generar dataframe

```python
Meses = ["Ene", "Feb","Mar","Abr","May","Jun","Jul","Ago","Sep","Oct","Nov","Dic"]
datos = pd.DataFrame({
    "Mes": Meses,
    "Ventas": np.random.randint(2000, 5000, 12),
    "Gastos": np.random.randint(1000, 4000, 12)
})

print(datos)
```

Graficos

```python
fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(2, 3)

#Crear los espacios para los subplots
ax1 = fig.add_subplot(gs[0, :2])  #Primer gráfico (barra, ocupando dos columnas)
ax2 = fig.add_subplot(gs[1, 0])   #Segundo gráfico (línea, una columna)
ax3 = fig.add_subplot(gs[:, 2])   #Tercer gráfico (dispersión, dos filas)

#Graficar los subplots

ax1.bar(datos["Mes"], datos["Ventas"], label="Ventas", color ="blue")
ax1.bar(datos["Mes"], datos["Gastos"], label="Gastos", color ="black")
ax1.set_title("Ventas y Gastos")
ax1.set_ylabel("Monto")
ax1.legend()

ax2.plot(datos["Mes"], datos["Gastos"], label="Gastos", color="green")
ax2.plot(datos["Mes"], datos["Ventas"], label="Ventas", color="gray")
ax2.set_title("Ventas y Gastos")
ax2.set_ylabel("Monto")
ax2.legend()

ax3.scatter(datos["Ventas"], datos["Gastos"], color = "green")
ax3.set_title("Ventas y Gastos")
ax3.set_xlabel("Ventas")
ax3.set_ylabel("Gastos")

plt.tight_layout()
plt.show()
```
