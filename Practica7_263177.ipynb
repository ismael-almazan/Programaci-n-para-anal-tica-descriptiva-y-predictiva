# **Carga y exploración del dataset:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset Iris de Seaborn
iris = sns.load_dataset('iris')

# Comprobar el dataset Iris
print(iris.head())

# Información general sobre los datos
print("\n Informacion General")
print(iris.info())

print("\n Estadisticas Descriptivas")
print(iris.describe())
```

# Gráfico de barras
*   Objetivo: Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.
*   Instrucciones: Utiliza un gráfico de barras para mostrar la media de las columnas petal_length y petal_width agrupadas por species.

```python
# calculo de proedio de ancho y largo de las especies de iris
promedios = iris.groupby('species')[['petal_length', 'petal_width']].mean()

# Cargar y agrupar datos
iris = sns.load_dataset('iris')
promedios = iris.groupby('species')[['petal_length', 'petal_width']].mean()

promedios.plot(kind="bar", color=["gold", "black"])
plt.ylabel("Media")
plt.xlabel("Especie")
plt.title("Media de ancho y largo de los petalos por especie")
plt.show()
```

# Histograma
*  Objetivo: Visualizar la distribución de los largos de los sépalos (sepal_length).
*  Instrucciones: Crea un histograma de la variable sepal_length y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título.

```python
#Grafico Histograma
plt.hist(iris['sepal_length'], bins=8, color='gold')
# Etiquetas
plt.xlabel('Largo del sepalo')
plt.ylabel('Frecuencia')
plt.title('Distribucion de la longitud de los sepalos')
plt.show()
```

# Gráfico de dispersión (scatter plot)
*  Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.
*  Instrucciones: Crea un gráfico de dispersión que compare petal_length con petal_width. Usa un color diferente para cada especie.

```python
# Grafico de dispersion
plt.scatter(iris['petal_length'], iris['petal_width'], c=iris['species'].map({'setosa': 'gold', 'versicolor': 'slategray', 'virginica': 'darkseagreen'}))
plt.xlabel('Largo del petalo')
plt.ylabel('Ancho del petalo')
plt.title('Relacion entre largo y ancho de los petalos')
plt.show()
```

# Gráfico de cajas (box plot)

*  Objetivo: Analizar la distribución de las medidas del sépalo (sepal_length, sepal_width).
*  Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables sepal_length y sepal_width por especie

```python

plt.boxplot([iris[iris['species'] == especie]['sepal_length'] for especie in iris['species'].unique()], labels=iris['species'].unique())
plt.title('Distribucion del largo de sepalo por especie')
plt.ylabel("Largo del sepalo")
plt.xlabel("Especie")
plt.show()

plt.boxplot([iris[iris['species'] == especie]['sepal_width'] for especie in iris['species'].unique()], labels=iris['species'].unique())
plt.title('Distribucion del ancho de sepalo por especie')
plt.ylabel("Ancho del sepalo")
plt.xlabel("Especie")
plt.show()
```

# Gráfico de líneas
*  Objetivo: Mostrar la tendencia de la longitud del pétalo (petal_length) a través de las observaciones.
*  Instrucciones: Crea un gráfico de líneas utilizando la variable petal_length (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

```python
plt.plot(iris.index, iris['petal_length'], color='gold')
plt.xlabel('Observacion')
plt.ylabel('Largo del petalo')
plt.title('Tendencia del largo del petalo a lo largo del dataset')
plt.show()
```
