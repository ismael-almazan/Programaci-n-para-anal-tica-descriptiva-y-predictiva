**Importar librerias**

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import pandas as pd
import seaborn as sns
```

Crear Dataframe

```python
datos = pd.DataFramedata = {
    "Normal": np.random.normal(loc=0, scale=1, size=1000),
    "Uniforme": np.random.uniform(low=0, high=1, size=1000),
    "Exponencial": np.random.exponential(scale=1, size=1000),
    "Poisson": np.random.poisson(3, 1000)
}
```

**Grafico**

```python
#Crear la figuro donde estaran contenidos los graficos de las distribuciones
fig, ax = plt.subplots(nrows=2, ncols=2)

#Crear las subfiguras
ax[0, 0].hist(datos["Normal"],      bins=20,  color="black", alpha=0.7)
ax[0, 1].hist(datos["Uniforme"],    bins=20,  color="red",   alpha=0.3)
ax[1, 0].hist(datos["Exponencial"], bins=20,  color="blue",  alpha=0.3)
ax[1, 1].hist(datos["Poisson"],     bins=20,  color="gray",  alpha=0.7)

#agregar los titulos
ax[0, 0].set_title("Normal")
ax[0, 1].set_title("Uniforme")
ax[1, 0].set_title("Exponencial")
ax[1, 1].set_title("Poisson")

#Ajustar la disposicion
plt.tight_layout()
plt.show()
```
