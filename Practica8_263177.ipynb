# Carga y exploración del dataset

```python
# Importar dataset
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = sns.load_dataset('tips')

#Exploracion del dataset tips
print("Primeros datos \n",df.head())
print("\nDescripcion de los datos o variables \n", df.describe())
print(df.info())
```

# Mapa de Calor (Heatmap)
*  Objetivo: Visualizar la correlación entre las variables numéricas en el dataset.
*  Instrucciones: Utiliza un mapa de calor para mostrar la matriz de correlación entre las variables numéricas (total_bill, tip, size).
*  Asegúrate de personalizar el mapa con una paleta de colores y etiquetas en los ejes.

```python
# Matriz de correlacion
corr = df[['total_bill', 'tip', 'size']].corr()

# Etiquetas
etiquetas = {"total_bill": "Tota cuenta", "tip": "Propina", "size": "# Personas"}

# Renombrar filas y columnas por la nuevas etiquetas en español
corr = corr.rename(index=etiquetas, columns=etiquetas)

# Graficar heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, cmap="magma", linewidths=0.5)

plt.title("Mapa de calor de correlaciones", fontsize=20)

```

# Diagrama de Violín
*  Objetivo: Mostrar la distribución de las propinas (tip) para diferentes días de la semana.
*  Instrucciones: Utiliza un diagrama de violín para comparar cómo varía la distribución de las propinas en cada uno de los días (day).
*  Personaliza el gráfico con colores y etiquetas adecuadas.

```python
#Utiliza un diagrama de violín para comparar cómo varía la distribución de las propinas en cada uno de los días (day)
plt.figure(figsize=(10,6))
violin = sns.violinplot(x="day", y="tip", data=df, palette="Set3")

#etiquetas
etiquetasdias= ['Jueves', 'Viernes', 'Sábado', 'Domingo']
violin.set_xticklabels(etiquetasdias)

#Leyendas
plt.xlabel("Dias")
plt.ylabel("Propina")
plt.title("Didstibucion de las propinas por dias de la semana")
```

# Gráfico de dispersión (Scatter Plot)
*  Objetivo: Analizar la relación entre el total de la cuenta (total_bill) y la propina (tip).
*  Instrucciones: Crea un gráfico de dispersión con total_bill en el eje X y tip en el eje Y. Usa un color diferente para cada día de la semana

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="day", palette="viridis")
#Etiquetas
plt.xlabel("Total cuenta")
plt.ylabel("Propina")
plt.title("Relacion entre el total de la cuenta y la propina")
```

# Gráfico de barras (Bar Plot)

*  Objetivo: Comparar el promedio de la propina (tip) por género (sex).
*  Instrucciones: Utiliza un gráfico de barras para mostrar el promedio de las propinas por género (hombres y mujeres).

```python
graficobarras= sns.barplot(data=df, x="sex", y="tip", estimator=np.mean, palette="pastel")

#Etiquetas

sexesp = ['Mujer', 'Hombre']
graficobarras.set_xticklabels(sexesp)

plt.xlabel("Genero")
plt.ylabel("Propina")
plt.title("Promedio de la propina por genero")

# Mostrar el gráfico
plt.show()
```
