```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats


#Importar del datset de AirQuality
df = pd.read_csv('/content/drive/MyDrive/Unidad02/AirQuality.csv', sep=';')
```

# **EDA**
Realizar un EDA completo, incluyendo:

* Descripción de los datos (medias, medianas, desviaciones estándar, etc.).

* Histogramas para las columnas numéricas.

* De la columna Time obten el día de la Semana y realiza una gráfica de barras.

* Boxplots para identificar outliers en las columnas numéricas.

* Matriz de correlación y mapa de calor.

* Pairplot para visualizar relaciones entre variables numéricas.

* Pruebas de normalidad (Shapiro-Wilk, Anderson-Darling, Kolmogorov-Smirnov) y QQplot para las columnas numéricas.

```python
#Comprobar que se cargaon los datos correctamente\
display(df.head())
display(df.dtypes)

#Utilizare display para poder imprimir mas de una tabla en el bloque de codigo

#Dimension de df
display(df.shape)
#Tiene dos columas vacias Unnamed: 16 y Unnamed: 15, se eliminan
df = df.drop(['Unnamed: 16', 'Unnamed: 15'], axis=1)

#Vambien se detecto que existen valores de -200 repetidamente en el datset lo cuales no son validos, tienen algun error que esta distrcionando los estadisticos y graficos
#asocioados con valores nulos o mediciones no validas

df = df.replace(-200, np.nan)

#Eliminar las filas donde todos los valores son NaN
df = df.dropna(how='all')

#Existen 366 registros donde aparentemente hay en muy mayoria datos perdidos o por erro de medicion casi la totalidad de las columnas
#Es conveniente eliminarlas ya que es muy poca la informacion que tiene, y para este caso especifico de estos registros no es covenientes imputar.

variableseliminar = ['PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'PT08.S3(NOx)', 'PT08.S4(NO2)', 'T', 'RH', 'AH']

# Eliminar las filas que tengan al menos un NaN en alguna de estas variables
df = df.dropna(subset=variableseliminar)


#Valores perdidos
display(df.isnull().sum())
display((df.isna().mean() * 100).round(2))


#NMHC(GT) mas del 90 % de los valores son perdidos, tambien se eliminara
df = df.drop('NMHC(GT)', axis=1)



# Descripción de los datos (medias, medianas, desviaciones estándar, etc.)
display(df.describe())

#Histogramas para las columnas numéricas.
#Seleccionar solo las columnas cnumerocas ya se con enteros o decimales
columns = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 12))
for i, col in enumerate(columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], kde=True, bins=30, color='teal')
    plt.title(f'Histograma de {col}', fontsize=10)
    plt.xlabel(col, fontsize=8)
    plt.ylabel('Frecuencia', fontsize=9)

plt.tight_layout()
plt.show()

#De la columna Time obten el día de la Semana y realiza una gráfica de barras
#Los dias de la semana no estan en time, sino en Date
#Date a datetime viene en DD/MM/YYY por lo que se tiene que especificar el formato
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

#Extraer el dia de la semana
df['DayWeek'] = df['Date'].dt.dayofweek
print(df['DayWeek'])
display(df.head())

#NO2(GT) por dia de la semana
plt.figure(figsize=(10, 6))
sns.barplot(x='DayWeek', y='NO2(GT)', data=df, palette='viridis')
plt.title('NO2(GT) por Dia de la Semana')
plt.xlabel('Dia de la Semana')
plt.ylabel('NO2(GT)')
plt.show()

#Boxplots para identificar outliers en las columnas numéricas
plt.figure(figsize=(15, 12))
for i, col in enumerate(columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[col], color='teal')

#Matriz de correlación y mapa de calor.
# Crear la matriz de correlación solo con las columnas numéricas
correlation_matrix = df[columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()

#Pairplot para visualizar relaciones entre variables numéricas.
sns.pairplot(df[columns])
plt.show()




#Pruebas de normalidad (Shapiro-Wilk, Anderson-Darling, Kolmogorov-Smirnov) y QQplot para las columnas numéricas.
#EL codigo fue tomado den notebook de Camputs Virtual
# Seleccionar columnas numéricas
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Función para realizar pruebas de normalidad
def normality_tests(data, column_name):
    data = data.dropna()  # Eliminar NaN
    if len(data) < 8:
        print(f"\nNo hay suficientes datos válidos en {column_name} para realizar las pruebas.")
        return
    if data.nunique() == 1:
        print(f"\nLa columna {column_name} tiene un solo valor constante, no se puede evaluar normalidad.")
        return

    print(f"\nPruebas de normalidad para la columna: {column_name}")

    # Shapiro-Wilk (muestra hasta 5000)
    shapiro_test = stats.shapiro(data.sample(min(5000, len(data)), random_state=42))
    print(f"Shapiro-Wilk Test: p-valor = {shapiro_test.pvalue:.5f}")

    # Kolmogorov-Smirnov
    ks_test = stats.kstest((data - data.mean()) / data.std(ddof=0), 'norm')
    print(f"Kolmogorov-Smirnov Test: p-valor = {ks_test.pvalue:.5f}")

    # Anderson-Darling
    anderson_test = stats.anderson(data, dist='norm')
    print(f"Anderson-Darling Test: Estadístico = {anderson_test.statistic:.5f}")
    for i in range(len(anderson_test.critical_values)):
        sl, cv = anderson_test.significance_level[i], anderson_test.critical_values[i]
        if anderson_test.statistic < cv:
            print(f"  Al nivel de significancia {sl}%, los datos parecen normales.")
        else:
            print(f"  Al nivel de significancia {sl}%, los datos NO parecen normales.")

    # D'Agostino-Pearson
    dagostino_test = stats.normaltest(data)
    print(f"D'Agostino-Pearson Test: p-valor = {dagostino_test.pvalue:.5f}")

# Función para graficar histograma y QQ-Plot
def plot_distribution(data, column_name):
    data = data.dropna()  # Ignorar NaN
    if len(data) == 0 or data.nunique() == 1:
        print(f"No se puede graficar {column_name} (sin datos o constante).")
        return

    plt.figure(figsize=(12, 5))

    # Histograma
    plt.subplot(1, 2, 1)
    sns.histplot(data, bins=50, kde=True, color="teal")
    plt.title(f"Distribución de {column_name}")

    # QQ-Plot
    plt.subplot(1, 2, 2)
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f"QQ-Plot de {column_name}")

    plt.show()

# Aplicar pruebas y gráficos a cada columna numérica
for column in numeric_columns:
    data = df[column].dropna()
    plot_distribution(data, column)
    normality_tests(data, column)



```
