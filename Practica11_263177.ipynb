```python
#Importar librerias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Cargar el Dataset de pengiins
df = sns.load_dataset("penguins")

print(df.info())



#crear figura de 2x2 para los subplots
fig, ax = plt.subplots(2,2)

#GRAFICOS

#Histograma de las longitudes de aletas.
sns.histplot(data=df, x="bill_length_mm", ax=ax[0, 0], color="skyblue")
ax[0, 0].set_title("Longitudes de aletas")
ax[0, 0].set_xlabel("Longitud")
ax[0, 0].set_ylabel("Frecuencia")

#Boxplot comparando la masa corporal por especie.
sns.boxplot(data=df, x="species", y="body_mass_g", ax=ax[0, 1],palette="viridis")
ax[0, 1].set_title("Masa corporal por especie.")
ax[0, 1].set_xlabel("Especie")
ax[0, 1].set_ylabel("Masa corporal")

#Dispersión entre masa corporal y largo de aletas
sns.scatterplot(data=df, x="bill_length_mm", y="body_mass_g", ax=ax[1, 0],color="darkgreen")
ax[1, 0].set_title("Masa corporal y largo de aletas")
ax[1, 0].set_xlabel("Largo de aletas")
ax[1, 0].set_ylabel("Masa corporal")

#Violin plot de la masa corporal según la especie.
sns.violinplot(data=df, x="species", y="body_mass_g", ax=ax[1, 1],palette="viridis")
ax[1, 1].set_title("Masa corporal según la especie")
ax[1, 1].set_xlabel("Especie")
ax[1, 1].set_ylabel("Masa corporal")



#Mostrar la figura
plt.tight_layout()
plt.show()



```
