# Ejercicio 01 Duplicidad Parcial Una columna: Contar y Filtrar los registros duplicados en Duration
*  Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.
*  Cuenta la cantidad de registros duplicados utilizando solo la información de columna Duration e imprime
*  Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.
*  Filtra los datos y obten los registros sin duplicados

```python
#Carga del conjunto de datos dirty.csv
df = pd.read_csv('/content/drive/MyDrive/Unidad02/dirtydata.csv')


#Verfificar su correcta carga
print(df.head())

#Contar la cantidad de registros duplicados utilizando solo la información de columna Duration e imprimir
duration_dup = df.duplicated(subset='Duration').sum()
print("\nCantidad de registros duplicados en Duration:", duration_dup)

#Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.
print("\nCantidad de registros que hay por cada valor único:")
print(df['Duration'].value_counts())

#Filtra los datos y obten los registros sin duplicados
df_sin_duplicados = df.drop_duplicates(subset='Duration')
print(df_sin_duplicados)
```

# Ejercicio 02 Duplicidad Parcial Dos columnaa: Contar y Filtrar los registros duplicados en Pulse y MaxPulse
*  Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.
*  Cuenta la cantidad de registros duplicados utilizando la información de las columnas Pulse y Maxpulse
*  Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.
*  Filtra los datos y obten los registros cuando keep = 'last'

```python
#Cuenta la cantidad de registros duplicados utilizando la información de las columnas Pulse y Maxpulse
pulse_maxpulse_dup = df.duplicated(subset=['Pulse', 'Maxpulse']).sum()
print("\nCantidad de registros duplicados en Pulse y Maxpulse:", pulse_maxpulse_dup)

#Ccantidad de registros que hay por cada valor único
print("\nCantidad de registros que hay por cada valor único:")
print(df[['Pulse', 'Maxpulse']].value_counts())

#Datos filtrados cuando Keep='last'
df_sin_duplicados1 = df.drop_duplicates(subset=['Pulse', 'Maxpulse'], keep='last')
print(df_sin_duplicados1)
```
